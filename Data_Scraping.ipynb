# install all the required libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import datetime
import os
from selenium import webdriver
import time
```

# setting a dynamic year range so it updates with every new season

```python
current_year = datetime.datetime.now().year
years = list(range(1956, current_year))
years
```

# Part 1: Scraping MVP data.

```python
# gets the link for all the years for MVP data and checks if it exists and if it does it skips to the next

for year in years:
    file_path = f"/Users/paramjaswal/Desktop/NBa/MVP/{year}.html"

    # Check if the file already exists
    if not os.path.exists(file_path):
        url = f'https://www.basketball-reference.com/awards/awards_{year}.html'
        data = requests.get(url)

        with open(file_path, "w+") as f:
            f.write(data.text)
            print(f"Data for {year} written to file.")
    else:
        print(f"Data for {year} already exists, skipping.")
```

```python
# we parse the data through beautiful soup finding the tag and saving it to mvp list

dfs = []
for year in years:
    with open(f"/Users/paramjaswal/Desktop/NBa/MVP/{year}.html", "r") as f:
        page = f.read()
        soup = BeautifulSoup(page, "html.parser")
        header_row = soup.find('tr', class_="over_header")
        if header_row:
            header_row.decompose()
        mvp_table = soup.find(id="mvp") or soup.find(id="nba_mvp")
        mvp = pd.read_html(str(mvp_table))[0]
        mvp["Year"] = year
        dfs.append(mvp)
```

```python
# we the eventually merge all the mvp years over axis = 0

mvps = pd.concat(dfs, ignore_index=True)
mvps
```

```python
# saving it to the local machine

mvps.to_csv("/Users/paramjaswal/Desktop/NBa/mvps.csv")
```

# Part 2: Scraping Players data

```python
# Initializing safari driver(you can initialize whatever browser driver you want)

driver = webdriver.Safari()
```

```python
# gets the link for all the years for Player data and checks if it exists and if it does it skips to the next using selenium

for year in years:
    # Check if the file already exists
    file_path = f"/Users/paramjaswal/Desktop/NBa/Player/{year}.html"

    if not os.path.exists(file_path):
        player_stats_url = f"https://www.basketball-reference.com/leagues/NBA_{year}_per_game.html"
        driver.get(player_stats_url)
        driver.execute_script("window.scrollTo(1, 10000)")
        with open(file_path, "w+") as f:
            f.write(driver.page_source)
            print(f"Data for {year} written to file.")
    else:
        print(f"Data for {year} already exists, skipping.")
```

```python
# close the safari driver

driver.quit()
```

```python
# we parse the data through beautiful soup finding the tag and saving it to player list

dfs = []
for year in years:
    with open(f"/Users/paramjaswal/Desktop/NBa/Player/{year}.html", "r") as f:
        page = f.read()
        soup = BeautifulSoup(page, "html.parser")

        header_rows = soup.find_all('tr', class_="thead")
        for header in header_rows:
            header.decompose()

        player_table = soup.find(id = "per_game_stats") or soup.find(id = "div_per_game_stats")
        player = pd.read_html(str(player_table))[0]
        player["Year"] = year
        dfs.append(player)
```

```python
# combining all the data

players = pd.concat(dfs, ignore_index=True)
players
```

```python
#  saving player data to the local machine

players.to_csv("/Users/paramjaswal/Desktop/NBa/players.csv")
```
