## Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import pickle
```

## Loading and Preparing Data

```python
stats = pd.read_csv("/Users/paramjaswal/Desktop/NBa/Player_mvp_stats.csv")
del stats["Unnamed: 0"]
stats
```

```python
stats.columns
```

### Splitting Data into Train and Test Sets

```python
predictors = ['Age',  'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'Year',
        'W', 'L', 'W/L%', 'GB', 'PS/G', 'PA/G', 'SRS']

target = 'Share'
```

## Model Training: Ridge Regression

```python
train_data = stats[stats['Year'] <= 2000]
```

## Model Training: Lasso Regression

```python
X = train_data[predictors]
y = train_data[target]
```

## Model Training: Support Vector Regression (SVR)

```python
# Ridge Regression
ridge_params = {'alpha': [0.01, 0.1, 1, 10, 100]}
ridge_grid = GridSearchCV(Ridge(), ridge_params, cv=5, scoring='neg_mean_squared_error')
ridge_grid.fit(X, y)
ridge_best = ridge_grid.best_estimator_
```

## Model Training: Random Forest

```python
# Lasso Regression
lasso_params = {'alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10]}
lasso_grid = GridSearchCV(Lasso(), lasso_params, cv=5, scoring='neg_mean_squared_error')
lasso_grid.fit(X, y)
lasso_best = lasso_grid.best_estimator_
```

## Model Training: Neural Network

```python
# SVR   
svm_params = {
    'C': [0.1, 1, 10],
    'epsilon': [0.01, 0.1, 1],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
}
svm_grid = GridSearchCV(SVR(), svm_params, cv=5, scoring='neg_mean_squared_error')
svm_grid.fit(X, y)
svm_best = svm_grid.best_estimator_
```

```python
# Random Forest
rf_params = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
rf_random = RandomizedSearchCV(RandomForestRegressor(), rf_params, n_iter=50, cv=5, scoring='neg_mean_squared_error', random_state=42, n_jobs=-1)
rf_random.fit(X, y)
rf_best = rf_random.best_estimator_
```

```python
# Neural Network Regression
nn_params = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'activation': ['relu', 'tanh', 'logistic'],
    'solver': ['adam', 'sgd'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate': ['constant', 'invscaling', 'adaptive']
}
nn_grid = GridSearchCV(MLPRegressor(max_iter=1000), nn_params, cv=5, scoring='neg_mean_squared_error')
nn_grid.fit(X, y)
nn_best = nn_grid.best_estimator_
```

## Model Evaluation and Comparison

### Saving the Best Performing Model

```python
ridge_best_params = ridge_grid.best_params_
ridge_best_model = Ridge(**ridge_best_params)
ridge_best_model.fit(X, y)
```

```python
best_lasso_params = lasso_grid.best_params_
lasso_final = Lasso(**best_lasso_params)
lasso_final.fit(X, y)
```

```python
best_svm_params = svm_grid.best_params_
svm_final = SVR(**best_svm_params)
svm_final.fit(X, y)
```

## Predicting MVP Share

```python
best_rf_params = rf_random.best_params_
rf_final = RandomForestRegressor(**best_rf_params)
rf_final.fit(X, y)
```

```python
best_nn_params = nn_grid.best_params_
nn_final = MLPRegressor(**best_nn_params)
nn_final.fit(X, y)
```

```python
X_test = stats[stats['Year'] == 2001][predictors]
y_test = stats[stats['Year'] == 2001]['Share']
```

```python
models = {
    'Ridge': ridge_best_model,
    'Lasso': lasso_final,
    'SVM' : svm_final,
    'Random Forest' : rf_final,
    'Neural Network Regression' : nn_final
}

mse_scores = {}

for model_name, model in models.items():
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_scores[model_name] = mse

mse_df = pd.DataFrame(list(mse_scores.items()), columns=['Model', 'MSE'])
mse_df = mse_df.sort_values(by='MSE')
```

```python
# Average Precision
```

```python
def avg_precision(combinations):
    actual = combinations.sort_values("Share", ascending=False).head(5)
    predicted = combinations.sort_values("Predictions", ascending=False)

    ps = []
    found = 0
    seen = 1
    for index, row in predicted.iterrows():
        if row["Player"] in actual["Player"].values:
            found += 1
            ps.append(found/seen)

        seen += 1
    return sum(ps)/len(ps)

avg_precision(combinations)
```

```python
def predict_mvp_share_with_names(player_stats, model, player_name="Hypothetical Player"):
    # Convert player stats to DataFrame


    stats = pd.read_csv("/Users/paramjaswal/Desktop/NBa/Player_mvp_stats.csv")
    del stats["Unnamed: 0"]

    predictors = ['Age',  'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'Year',
        'W', 'L', 'W/L%', 'GB', 'PS/G', 'PA/G', 'SRS']


    player_data = pd.DataFrame([player_stats])
    
    # Extract relevant data from the original stats DataFrame based on the provided year
    data_for_year = stats[stats['Year'] == player_data['Year'][0]]
    
    # Extract player names for the given year and append the hypothetical player's name
    player_names = data_for_year['Player'].tolist() + [player_name]
    
    # Extract only predictor columns and concatenate with hypothetical player data
    data_for_prediction = pd.concat([data_for_year[predictors], player_data], ignore_index=True)
    
    # Predict MVP share for all players including the hypothetical player
    predictions = model.predict(data_for_prediction)
    
    # Create the final DataFrame with player names and their predicted MVP share
    result_df = pd.DataFrame({
        'Player': player_names,
        'Predicted MVP Share': predictions
    })
    
    sorted_df = result_df.sort_values(by="Predicted MVP Share", ascending=False).reset_index(drop=True)
    sorted_df.index = sorted_df.index + 1

    # Check if the player exists in the DataFrame
    if player_name in sorted_df['Player'].values:
        player_index = sorted_df[sorted_df['Player'] == player_name].index[0]
        print(f"Rank of your player {player_name} is {player_index}")
    else:
        print(f"Player {player_name} not found in the predictions.")
        
    return sorted_df
```
